```python
from langchain_groq import ChatGroq
from langgraph.graph import StateGraph , START , END
from typing import TypedDict
```

```python
from google.colab import userdata
api_key = userdata.get('groq_api_key')
# print(api_key)
```

```python
model = ChatGroq(
    model="llama-3.1-8b-instant",
    api_key=api_key
    )
```

```python
class qn_state(TypedDict):
  question:str
  answer:str

def qn_funct(state:qn_state):
  quest = state["question"]

  prompt = f"Answer the following the Question in only one word, {quest}"

  answer = model.invoke(prompt).content

  state["answer"] = answer
  return state
```

```python
graph = StateGraph(qn_state)

graph.add_node("question_answer" ,  qn_funct)

graph.add_edge(START , "question_answer")
graph.add_edge("question_answer" , END)
```

```python
workflow = graph.compile()
```

```python
response = workflow.invoke({"question":"What is the Capital of Pakistan?"})
print(response["answer"])
```
